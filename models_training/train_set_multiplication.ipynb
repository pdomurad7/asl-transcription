```python
import pandas as pd
import numpy as np
```

```python
# Definiujemy źródło danych oryginalnych
SOURCE_FILE = 'datasets/pawel_video_dataset.csv'

# Definiujemy ścieżkę do zapisu danych augmentowanych
TARGET_FILE = 'datasets/pawel_video_dataset_multiplications/pawel_video_dataset_moved_221_69.csv'

# Definiujemy maksymalne losowe przesunięcie w osiach x, y, z
przesuniecie_x_max = 0.02
przesuniecie_y_max = 0.02
przesuniecie_z_max = 0.01
```

```python
df = pd.read_csv(SOURCE_FILE)
original_df_head = df.head()
original_df_head
```

```python
for i in range(21):
    for coord in ['x', 'y', 'z']:
        col_name = f'{coord}{i:02d}'
        if coord == 'z':
            przesuniecie_max = przesuniecie_z_max
        elif coord == 'y':
            przesuniecie_max = przesuniecie_y_max
        else:
            przesuniecie_max = przesuniecie_x_max
        
        # Losujemy przesunięcie dla całej kolumny zawierającej współrzędne jednego z punktów charakterystycznych
        przesuniecie = np.random.uniform(-przesuniecie_max, przesuniecie_max, df.shape[0])
        
        # Dodajemy przesunięcie do współrzędnych
        df[col_name] += przesuniecie
df.head()
```

```python
df.to_csv(TARGET_FILE, index=False)
```
